```python
import pandas as pd
import matplotlib.pyplot as plt

def analyze_earth_meteorites_data(file_path):
    try:
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)

        # Convert the 'year' column to datetime
        df['year'] = pd.to_datetime(df['year'], errors='coerce')

        # Get all the Earth meteorites that fell before the year 2000
        meteorites_before_2000 = df[df['year'] < pd.to_datetime('2000-01-01')]
        print("Earth meteorites that fell before the year 2000:")
        print(meteorites_before_2000)

        # Get all the earth meteorites coordinates that fell before the year 1970
        meteorites_coordinates_before_1970 = df[df['year'] < pd.to_datetime('1970-01-01')][['reclat', 'recclong']]
        print("Earth meteorites coordinates that fell before the year 1970:")
        print(meteorites_coordinates_before_1970)

        # Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more than 10000kg
        meteorites_mass_more_than_10000kg = df[df['mass'] > 10000]
        print("Earth meteorites with mass more than 10000kg:")
        print(meteorites_mass_more_than_10000kg)

        # Visualize the analysis using plots
        # Plot 1: Histogram of meteorite years that fell before the year 2000
        plt.figure(figsize=(10, 6))
        plt.hist(meteorites_before_2000['year'], bins=40, edgecolor='black')
        plt.xlabel('Year')
        plt.ylabel('Frequency')
        plt.title('Distribution of Earth Meteorite Years')
        plt.show()

        # Plot 2: Scatter plot of meteorite coordinates before 1970
        plt.figure(figsize=(10, 6))
        plt.scatter(meteorites_coordinates_before_1970['recclong'], meteorites_coordinates_before_1970['reclat'], c='blue', alpha=0.5)
        plt.xlabel('Longitude')
        plt.ylabel('Latitude')
        plt.title('Earth Meteorite Coordinates before 1970')
        plt.show()
        
        # Plot 3: Bar plot of earth meteorites with mass greater than 10000kg
        plt.figure(figsize=(10, 6))
        plt.bar(meteorites_mass_more_than_10000kg['Name of Earth Meteorite'], meteorites_mass_more_than_10000kg['mass'])
        plt.xlabel('Meteorite')
        plt.ylabel('Mass (kg)')
        plt.title('Earth Meteorites with Mass > 10000kg')
        plt.xticks(rotation=90)
        plt.ylim(ymin=10000)  # Set the y-axis lower limit to 10000
        plt.show()

    except Exception as e:
        print("An error occurred:", str(e))


# Specify the file path of the Earth meteorites data CSV file
file_path = 'earth_meteorites.csv'

# Call the function to analyze the Earth meteorites data
analyze_earth_meteorites_data(file_path)
```
